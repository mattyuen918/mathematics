```python
import numpy as np
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from ipywidgets import widgets, interact
from bokeh.plotting import ColumnDataSource
from bokeh.models import HoverTool
```

```python
def fun(a):
    theta = np.linspace(-3,6,1000)
    x = 4*np.cos(theta*a) + 2*np.cos(2*theta*a)
    y = 4*np.sin(theta*a) - 2*np.sin(2*theta*a)
    output_notebook()
    fig = figure(title = 'Exercise 1')
    fig.line(theta,x,y)
    show(fig)
interact(fun,a=widgets.IntSlider(min= 2, max = 15, step = 1))
```

```python
output_notebook()
source = ColumnDataSource(
        data=dict(
            x=[1, 3, 2],
            y=[0, 0, 2],
            desc=['A', 'B', 'C', 'D'],
        )
    )

hover = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("(x,y)", "($x, $y)"),
            ("desc", "@desc"),
        ]
    )
p = figure(plot_width=400, plot_height=400, tools=[hover],
           title="Exercise 2")

p.circle('x', 'y', size=20, source=source)

show(p)
```
