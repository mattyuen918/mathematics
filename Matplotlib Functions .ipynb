# Plotting Functions in Python using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

Plot the [Figure Eight Curve](http://www-groups.dcs.st-and.ac.uk/history/Curves/Eight.html)

$$ x = \sin(t), y = \sin(t)\cos(t) \space \forall t \in [0,2\pi] $$ 

```python
t = np.linspace(0,2*np.pi,1000)
x = np.sin(t)
y = np.sin(t)*np.cos(t)
plt.plot(x,y)
```

Plot the [Butterfly Curve](https://en.wikipedia.org/wiki/Butterfly_curve_(transcendental)

```python
t = np.linspace(0,12*np.pi,1000)
x = np.sin(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-np.sin(t/12)**5)
y = np.cos(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-np.sin(t/12)**5)
plt.plot(x,y),plt.axis([-4,4,-4,4]);
```

Plot the surface: $$ z = x^2-y^2 \space \text{for x,y}\in[-2,2] $$ 

```python
plt.figure()
ax = plt.axes(projection = '3d')
x_bound = np.linspace(-2,2,1000)
y_bound = np.linspace(-2,2,1000)
x,y = np.meshgrid(x_bound,y_bound)
z = x**2 - y**2 
ax.plot_surface(x,y,z)
```
