# Slope Field Example 

Let's write a function called slope_field which takes 4 input parameters: f, tlims, ylims, grid_step. 

The function of two variables f represents the right side of a first order ODE equation. 

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
def f(t,y):
    return y**2-t**2 
```

```python
def slope_field(f,tlims,ylims,grid_step):
    " f is the right side of first order ODE,
    " tlims and ylims are display limits of the figure, and grid_step is the distance between grid points and on the plot. "
    t = np.arange(tlims[0],tlims[1],grid_step)
    y = np.arange(ylims[0],ylims[1],grid_step)
    for T in t: 
        for Y in y: 
            small = grid_step/10
            slope = f(T,Y)
            domain = np.linspace(T-small,T+small,2)
            z = slope*(domain-T)+Y
            plt.plot(domain,z,'b'),plt.tight_layout()
```

```python
slope_field(f,[-3,3],[-2,2],0.2)
```
